```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker


df.columns = df.columns.str.strip()


food_col = 'which food you prefer'


food_counts = df[food_col].value_counts().sort_values(ascending=False)


plt.figure(figsize=(8, 6))
plt.bar(food_counts.index, food_counts.values, color='darkorange')

plt.title(f'Distribution of Food Preference', fontsize=16)
plt.xlabel('Food Preference', fontsize=12)
plt.ylabel('Number of Respondents', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')


plt.gca().yaxis.set_major_locator(ticker.MaxNLocator(integer=True))
plt.tight_layout()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker


df.columns = df.columns.str.strip()


preference_col = 'Do you prefer coffee or tea?'


preference_counts = df[preference_col].value_counts().sort_values(ascending=True)

# Create the figure and axis
plt.figure(figsize=(8, 5))

# Create a horizontal bar chart (barh)
bars = plt.barh(preference_counts.index, preference_counts.values, color=['sienna', 'gold'])

plt.title('Coffee vs. Tea Preference', fontsize=16)
plt.xlabel('Number of Respondents', fontsize=12)
plt.ylabel('Preference', fontsize=12)
plt.grid(axis='x', linestyle='--')

# Ensure x-axis shows integers for the count
plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(integer=True))

# Add value labels to the bars
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2,
             f'{bar.get_width()}',
             va='center', ha='left', fontsize=10)

plt.tight_layout() # Adjust layout
```
